# How to fetch data from an API

```python
#Source: TMDB wesbite
#https://developers.themoviedb.org/3/movies/get-top-rated-movies
```

```python
import pandas as pd
```

```python
import requests
```

```python
#Response[200] indicates that the link is working fine and we can move ahead with the procedure.
response=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=75b814fdc3c1a18bface7bd7059718c3&language=en-US&page=1")
response
```

```python
#to convert the response to json format
response.json()
```

```python
#to get a list
response.json()['results']
```

```python
#to convert the list to a dataframe
pd.DataFrame(response.json()['results'])[['id','original_title','title','popularity','release_date','vote_average','vote_count']]
```

```python
#creating an empty dataframe so that I can append all the data continuously to one dataframe.
df=pd.DataFrame()
df
```

```python
#We have successfully fetched the data from an api single page. Now, we need to repeatedly do the same procedure to get data from all the pages.
#Every loop will fetch a page from the website in the form of a dataframe.
#To know how many loops we need to run, just go to "http://jsonviewer.stack.hu/" and paste the json data and click on viewer. It will output the number of pages, total_results.

for i in range(1,459):
    response=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=75b814fdc3c1a18bface7bd7059718c3&language=en-US&page={}".format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','original_title','title','popularity','release_date','vote_average','vote_count']]
    df=df.append(temp_df, ignore_index=True)
    df.to_csv('Movies_rating.csv')
```
